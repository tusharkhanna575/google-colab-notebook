```python
#Modules Imported
import pandas as pd
import numpy as np
import ast
import pickle
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
songs=pd.read_csv("/content/ex.csv", encoding= 'unicode_escape')
```

```python
songs.head()
```

```python
new_columns = {"Song-Name": "name", "Singer/Artists": "singer", "Album/Movie":"movie", "User-Rating":"rating", "Genre":"genre"}
songs = songs.rename(columns=new_columns)
```

```python
songs.head()
```

```python
songs.info()
```

```python
# dropping null values
songs.dropna(inplace=True)
```

```python
songs['tags'] = songs['singer'] + " " + songs['genre'] + " " + songs['movie']
songs.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
# converting each song to vector format
vector = cv.fit_transform(songs['tags']).toarray()
vector.shape
```

```python
# generates a list of 5000 values
cv.get_feature_names()
# removing redundant names
```

```python
# taking cosine distance between two vectors i.e. songs
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
similarity
```

```python
songs['name']
```

```python
songs[songs['name']=='Coca Cola'].index[0]
print(type(songs))
```

```python
# recommend fuction to generate songs
def recommend(song):
    index = songs[songs['name'] == song].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        #print(songs.iloc[i[0]])
        print(songs['name'][songs.iloc[i[0]].name])
```

```python
recommend("Mungda")
```

```python
# Python pickle module is used for serializing and de-serializing python object structures.
import pickle
pickle.dump(songs,open('song_list.pkl','wb'))
pickle.dump(similarity,open('similarity.pkl','wb'))
```
